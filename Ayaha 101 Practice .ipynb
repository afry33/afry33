```python
import pandas as pd
import requests
from io import StringIO

# URL containing the GME revenue data
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Set up a User-Agent header to mimic a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Send a GET request with the User-Agent header
response = requests.get(url, headers=headers)

# If the request was successful (status code 200), proceed to read the table
if response.status_code == 200:
    # Wrap the HTML content in StringIO to avoid the FutureWarning
    html_content = StringIO(response.text)
    
    # Read the revenue data from the HTML response
    gme_revenue = pd.read_html(html_content)[0]

    # Display the last five rows of the GME revenue data
    print(gme_revenue.tail())
else:
    print(f"Failed to retrieve data. HTTP status code: {response.status_code}")
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Get Tesla stock data for the past year
tesla_data = yf.download('TSLA', period='1y')

# Function to make the graph
def make_graph(data):
    # Plotting Tesla's stock closing price
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='Tesla Stock Price', color='blue')
    
    # Adding title and labels
    plt.title('Tesla Stock Price Over the Last Year')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    
    # Show the legend
    plt.legend()

    # Display the plot
    plt.show()

# Call the function to show the graph
make_graph(tesla_data)
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Get GameStop stock data for the past year
gamestop_data = yf.download('GME', period='1y')

# Function to make the graph
def make_graph(data):
    # Plotting GameStop's stock closing price
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='GameStop Stock Price', color='red')
    
    # Adding title and labels
    plt.title('GameStop Stock Price Over the Last Year')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    
    # Show the legend
    plt.legend()

    # Display the plot
    plt.show()

# Call the function to show the graph
make_graph(gamestop_data)
```
